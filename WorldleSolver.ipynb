```python
# Worldle Solver
# World Solver Plays Wordle
import random


WORD_LENGTH = 5
# options_to_print = 5
num_games_to_play = 25

alphabet = [chr(x) for x in range(97, 123)]
# print(alphabet)

def word_score(word, dictionary):
  letter_scores = {}
  for letter in alphabet:
    letter_scores[letter] = 0
  
  for d in dictionary:
    for letter in d:
      letter_scores[letter] += 1
  
  return sum([letter_scores[letter] for letter in word])

def color_word(guess, correct):

  color_to_return = ''
  frequency_of_letters = {}

  for char in correct:
    if char in frequency_of_letters:
      frequency_of_letters[char] += 1
    else:
      frequency_of_letters[char] = 1


  for i, char in enumerate(guess):
    if char == correct[i]:
      color_to_return += 'G'
    elif char in frequency_of_letters and frequency_of_letters[char] > 0:
      frequency_of_letters[char] -= 1
      color_to_return += 'Y'
    else:
      color_to_return += 'B'

  return color_to_return

# Load in a Dictionary
dictionary = []

total_attempts = 0

for i in range(num_games_to_play):
  with open('./drive/MyDrive/five_letter_dic', 'r') as f:
    lines = f.readlines()
    # print(lines[:10])
    dictionary = [line.strip() for line in lines]
  # Sort dictionary where words with more letters show up first...
  dictionary.sort(key = lambda x : word_score(x, dictionary), reverse = True)
  # print(dictionary[:10])

  word_to_guess = random.choice(dictionary)

  good_letters = [set(alphabet) for i in range(WORD_LENGTH)]

  solved = False

  attempts = 0

  while (not solved):
    
    attempts += 1

    cur_yellow_letters = set()  
    solved = True

    # print("Enter a word that you guessed.")
    # Enter a word
    # word = input()
    word = dictionary[0]

    # print("Enter the colors that you received. e.g. BGBBY")
    # Enter the colors for the letters
    colors = color_word(word, word_to_guess)

    for index, color in enumerate(colors):
        if color == 'G':
            good_letters[index] = set([word[index]])
        elif color == 'Y':
            cur_yellow_letters.add(word[index])
            good_letters[index].remove(word[index])
            solved = False
        else:
            solved = False
            if word[index] not in cur_yellow_letters:
              for i in range(len(good_letters)):
                  if len(good_letters[i]) != 1:
                      if word[index] in good_letters[i]:
                          good_letters[i].remove(word[index])

    dictionary = list(filter(lambda x: all([x[i] in good_letters[i] 
                        for i in range(len(good_letters))]+
                        [letter in x for letter in cur_yellow_letters]), dictionary))
    dictionary.sort(key = lambda x : word_score(x, dictionary), reverse = True)
    # for i in range(min(options_to_print, len(dictionary))):
    #   print(dictionary[i])
  total_attempts += attempts
  print(f"We were able to guess {word_to_guess} in {attempts} guesses!")

print(f"On average, we took {total_attempts/num_games_to_play} guesses to"+
      " get the word correct!")
```
